```python
import pandas
import csv
import unicodedata
import re
```

```python
test = list(csv.reader(open('data/Q1-recom-elno_cleaned_data_current/test.csv', 'rt')))
```

```python
test[1]
```

```python
def unicodeToAscii(s):
    return ''.join(
        c for c in unicodedata.normalize('NFD', s)
        if unicodedata.category(c) != 'Mn'
    )

def normalizeString(s):
    s = unicodeToAscii(s.lower().strip())
    s = re.sub(r"([.!?])", r" \1", s)
    s = re.sub(r"[^a-zA-Z.!?]+", r" ", s)
    s = re.sub(r"\s+", r" ", s).strip()
    return s
```

```python
target = []
for l in test[1:]:
    target.append(normalizeString(l[-1]))
```

```python
len(target)
```

```python
len(test)
```

```python
with open('data/Q1-recom-elno_cleaned_data_current/test_target.csv', 'wt') as f:
    for l in target:
        f.write('{}\n'.format(l))
```

```python
text = []
for l in test[1:]:
    text.append(normalizeString(' '.join(l[:-1])))
```
